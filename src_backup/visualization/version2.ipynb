```python
import holoviews as hv
import param, parambokeh

from bokeh.document import Document
from holoviews.operation.datashader import datashade
from holoviews.streams import RangeXY

import dask.array
import numpy as np
import h5py

hv.extension('bokeh')
```

# Load data

```python
with h5py.File(r'C:/Users/haoyu/Documents/Python Scripts/diffusion_map_dev/testSample.h5') as h5file:
    keys = h5file.keys()
    holder = [h5file[name] for name in keys]
    data = np.array(dask.array.concatenate(holder))

print(data.shape)
```

# Generate some indexes

```python
x = np.random.rand(10**7)*100
y = np.random.rand(10**7)*100
idx = np.random.randint(0, high = data.shape[0], size=10**7)

# Extract some of the positions and indexes
num_to_show = 900
points = hv.Points(np.stack((x[:num_to_show], y[:num_to_show]),axis=1))
index_map = idx[:num_to_show]
```

# Visualization Step 2: Box selection

```python
# Declare points as source of selection stream
selection = hv.streams.Selection1D(source=points)

# Write function that uses the selection indices to slice points and compute stats
def get_global_index(index):
    selected = points.iloc[index]
    if index:
        return hv.Points(selected)
    else:
        return hv.Points(np.random.rand(100,2 ) * 100)
```

```python
%%opts Points [tools=['box_select', 'lasso_select']]

shaded = datashade(hv.Scatter((x,y)), dynamic=True)

shaded*points.options({'Points': {'color': 'red', 'size': 3}}) + hv.DynamicMap(get_global_index, streams=[selection])
```
